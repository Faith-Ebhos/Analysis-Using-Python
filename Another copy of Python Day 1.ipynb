```python
# What if user entry is decimal number?
print("Tell me time in hours and I will calculate in seconds")
time_hour = input("how many hours? ")
print(f"Time in seconds is {int(time_hour) * 60 * 60}")
# int function cant conver decimal to integer
```

```python
# What if user entry is decimal number?
print("Tell me time in hours and I will calculate in seconds")
time_hour = input("how many hours? ")

print(f"Time in seconds is {float(time_hour) * 60 * 60}")
# float function can convert decimal
```
